# Neural AntMaze — Full Training (Colab)

Trains a hierarchical neural SF agent on AntMaze UMaze — an 8-DOF Ant
quadruped navigating through a U-shaped maze to reach a goal.

Architecture:
- ActionConditionedSFNetwork (6561 discrete actions = 3^8 joint torques)
- HierarchicalNeuralSRAgent with 4-cluster macro planning
- Two-phase training with phase-boundary management

**Runtime**: ~2-4 hours on GPU, longer on CPU

```python
# 2. Verify imports and AntMaze environment
import sys
sys.path.insert(0, '.')

import torch
import gymnasium as gym
import gymnasium_robotics
import numpy as np

gym.register_envs(gymnasium_robotics)

# Test AntMaze environment
env = gym.make('AntMaze_UMaze-v5')
obs, _ = env.reset()
print(f'Observation keys: {list(obs.keys())}')
print(f'observation shape: {obs["observation"].shape}')
print(f'achieved_goal shape: {obs["achieved_goal"].shape}')
print(f'desired_goal shape: {obs["desired_goal"].shape}')
print(f'Action space: {env.action_space}')
print(f'Flattened obs dim: {obs["observation"].shape[0] + 2 + 2}')
env.close()

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f'\nUsing device: {device}')
print(f'PyTorch: {torch.__version__}')
```

```python
# 4. Inspect training diagnostics
import torch
import numpy as np

checkpoint = torch.load('data/neural_ant_maze/checkpoint.pt',
                        map_location='cpu', weights_only=False)
print('Checkpoint keys:', list(checkpoint.keys()))
print(f'Total steps: {checkpoint["total_steps"]}')
print(f'Final epsilon: {checkpoint["epsilon"]:.4f}')
```

```python
# 5. Plot training curves
import matplotlib.pyplot as plt
import numpy as np

log = checkpoint.get('training_log', {})

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for ax, (key, title) in zip(axes.flat, [
    ('sf_loss', 'SF TD Loss'),
    ('reward_loss', 'Reward Prediction Loss'),
    ('episode_reward', 'Episode Reward'),
    ('episode_steps', 'Episode Steps'),
]):
    data = log.get(key, [])
    if data:
        ax.plot(data, alpha=0.3, linewidth=0.5)
        w = min(100, len(data) // 10 + 1)
        if w > 1:
            sm = np.convolve(data, np.ones(w)/w, mode='valid')
            ax.plot(np.arange(w-1, w-1+len(sm)), sm, linewidth=2)
    ax.set_title(title)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
